## ë°°ì—´ì˜ ë³µì‚¬
* copyì™€ viewì˜ ì°¨ì´ì ì€ copyëŠ” ìƒˆ ë°°ì—´(ê¹Šì€ ë³µì‚¬)ì´ê³  viewëŠ” ì›ë˜ ë°°ì—´ê³¼ ì—°ê²° ë˜ì–´ ìˆë‹¤ëŠ” ê²ƒ(ì–•ì€ ë³µì‚¬)

* viewëŠ” ë°ì´í„°ë¥¼ ì†Œìœ í•˜ì§€ ì•Šìœ¼ë©° viewì— ëŒ€í•œ ëª¨ë“  ë³€ê²½ ì‚¬í•­ì€ ì›ë˜ ë°°ì—´ì— ì˜í–¥ì„ ë¯¸ì¹˜ê³  ì›ë˜ ë°°ì—´ì— ëŒ€í•œ ëª¨ë“  ë³€ê²½ ì‚¬í•­ì€ ë³´ê¸°ì— ì˜í–¥ì„ ì¤Œ

````python
arr = np.array(range(1,6))
x = arr.copy()
arr[0] = 40
print(f"arr : {arr}")
print(f"x: {x}")
````

````python
arr = np.array(range(1,6))
x = arr.view()
arr[0] = 40
print(f"arr : {arr}")
print(f"x: {x}")
````

````python
import numpy as np
arr = np.array(range(1,6))

x = arr.copy()
y = arr.view()

# ë°°ì—´ ë°ì´í„° ì†Œìœ  ì—¬ë¶€ ë°˜í™˜
print(x.base) # xëŠ”(copy) ê¹Šì€ ë³µì‚¬
print(y.base) # yëŠ”(view) ì–•ì€ ë³µì‚¬ 
````
