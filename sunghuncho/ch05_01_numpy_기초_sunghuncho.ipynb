|ìë£Œí˜•|íŠ¹ì§•|

|ë¦¬ìŠ¤íŠ¸(list)|ê°ê°ì˜ ì›ì†Œê°€ ë‹¤ë¥¸ ìë£Œí˜•ì´ ë  ìˆ˜ ìˆìŒ|
|ë°°ì—´(array)|ì—°ì†ì ì¸ ë©”ëª¨ë¦¬ ë°°ì¹˜ë¥¼ ê°€ì§€ê¸° ë•Œë¬¸ì— ëª¨ë“  ì›ì†Œê°€ ê°™ì€ ìë£Œí˜•ì´ì–´ì•¼ í•¨<br>ì›ì†Œì— ëŒ€í•œ ì ‘ê·¼ê³¼ ë°˜ë³µë¬¸ ì‹¤í–‰ì´ ë¹¨ë¼ì§|

````python
import random

def average_python(n):
  s=0
  for i in range(n):
    s+=random.random()
  return s/n
````

````python
%time average_python(100_000)
````

````python
def average_numpy(n):
  s=np.random.random(n)
  return s.mean()
````

````python
%time average_numpy(100_000)
````
