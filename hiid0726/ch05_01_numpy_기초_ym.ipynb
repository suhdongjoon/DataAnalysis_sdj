## ndarray
* ë°°ì—´ ê°ì²´ íƒ€ì…(ìë£Œí˜•)
* ë¦¬ìŠ¤íŠ¸ì™€ ë™ì¼í•´ ë³´ì´ì§€ë§Œ ì°¨ì´ê°€ ë§ìŒ

|ìë£Œí˜•|íŠ¹ì§•|
|:-|:-|
|ë¦¬ìŠ¤íŠ¸(list)|ê°ê°ì˜ ì›ì†Œê°€ ë‹¤ë¥¸ ìë£Œí˜•ì´ ë  ìˆ˜ ìˆìŒ|
|ë°°ì—´(array)|ì—°ì†ì ì¸ ë©”ëª¨ë¦¬ ë°°ì¹˜ë¥¼ ê°€ì§€ê¸° ë•Œë¬¸ì— ëª¨ë“  ì›ì†Œê°€ ê°™ì€ ìë£Œí˜•ì´ì–´ì•¼ í•¨<br>ì›ì†Œì— ëŒ€í•œ ì ‘ê·¼ê³¼ ë°˜ë³µë¬¸ ì‹¤í–‰ì´ ë¹¨ë¼ì§|

````python
import random;


def average_python(n):
# nê°œì˜ ëœë¤í•œ 0~1ì˜ ì‹¤ìˆ˜ë¥¼ í‰ê· í•´ì£¼ëŠ” í•¨ìˆ˜
  s=0;
  for i in range(n) :
    s += random.random()
  return s/n;
````

````python
%time average_python(100000)
````

````python
%timeit average_python(10)
````

````python
def average_numpy(n):
  # nê°œì˜ ëœë¤í•œ 0~1ì˜ ì‹¤ìˆ˜ë¥¼ numpyë¥¼ ì‚¬ìš©í•´ì„œ í‰ê· ë‚´ëŠ” í•¨ìˆ˜
  s = np.random.random(n) # numpyì— ìˆëŠ” random í•¨ìˆ˜
  return s.mean() #.mean() = í‰ê· 
````

````python
%time average_numpy(10)
````

````python
%timeit average_numpy(10)
````

````python
arr = np.array(range(1,6))
x = arr.view()   ###### ì—°ê²°ë˜ëŠ” ê²ƒ!! (ì–•ì€ ë³µì‚¬)
arr[0] = 5 ## xë„ ê°™ì´ ë°”ë€ë‹¤!!! (ì–•ì€ ë³µì‚¬)
x[1] = 10
print(arr)
print(x)
````
